## 1. Buatlah program BMI menggunakan bahasa Python

- BMI (Body Mass Index) merupakan perhitungan berat badan ideal seseorang dimana rumus dari BMI adalah sebagai berikut:
    
    ![rumus-bmi-cara-menghitung-berat-badan-ideal.jpeg](https://openaccesspub.org/article/926/images/image1.jpg)
    
- Buatlah program yang menerima 2 input (berat badan & tinggi badan) kemudian keluarkan hasil BMI nya dan keterangan dari BMI tersebut

```python
#Tulis Program Disini
weight = int(input('Input your body weight: '))
height = int(input('Input your body height: '))

bmi = weight/(height**2)

print(f"your BMI is: {bmi}")

if bmi < 18.5:
  print('you\'re underweight')
elif bmi < 24.9:
  print('you\'re normal')
elif bmi < 29.9:
  print('you\'re overweight')
else:
  print('you\'re very overweight')
```

### 2. Buatlah Piramida Bintang menggunakan bahasa Python

- Buatlah program yang dapat mencetak output seperti dibawah ini menggunakan looping python
    
    ![piramida-bintang.png](https://i.ibb.co/BtqjZrL/piramida-bintang.png)
    

```python
#Tulis Program Disini
import math

height = int(input('Enter your pyramid height (odd number): '))

i = 0
for i in range(height):
  for j in range(math.floor((height * 2 - 1)/2) - i):
    print(" ", end="")

  for k in range(2 * i + 1):
    print("*", end="")

  print()
```
