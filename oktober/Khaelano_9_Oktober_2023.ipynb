```python
def func2(x):
  return (x**2 - 1)/(x - 1)

input = (2, 1.05, 1.001, 1.00000017, 0.5, 0.99, 0.999975, 0.9999999)

for i in input:
  print(func2(i))
```

```python


x = np.arange(-3, 6, 0.5)
y = func2(x)
plt.plot(x, y, "-r", label="y=(x^2-1)/(x-1)")
plt.title("Grafik y=(x^2-1)/(x-1)")


plt.xlabel("x", color='#1C2833')
plt.ylabel("t", color='#1C2833')
plt.legend(loc='upper left')
plt.show()
```

```python
def func3(x):
  return 1/x**2

input = (1, 0.5, 0.01, 0.0001, 0.000005, -1, -0.5, -0.01, -0.0001, -0.000005)

for i in input:
  print(func3(i))
```

```python
x = np.arange(0, 0.5, 0.01)
y = func3(x)
plt.plot(x, y, "-r", label="y=1/x^2")
plt.title("Grafik y=1/x^2")


plt.xlabel("x", color='#1C2833')
plt.ylabel("t", color='#1C2833')
plt.legend(loc='upper left')
plt.show()
```
