```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def fungsi1(x):
  return 1/x
```

```python
input = np.arange(0, 5, 0.1)
output = fungsi1(input)

plt.plot(input, output, label="1/x")
plt.title("grafik 1/x")
```

```python
input = np.arange(-5, 5, 0.1)
output = fungsi1(input)

plt.plot(input, output, label="1/x")
plt.title("grafik 1/x")
```

```python
input = np.arange(-5, 5, 1)
output = fungsi1(input)

plt.plot(input, output, label="1/x")
plt.title("grafik 1/x")
```

```python
def fungsi2(x):
  return abs(x)/x

input = np.arange(-5, 5, 0.1)
output = fungsi2(input)

plt.plot(input, output, label="1/x")
plt.title("grafik 1/x")
```
