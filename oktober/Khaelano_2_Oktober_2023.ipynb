```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
```

```python
def kuadrat(x):
  return x**2
```

```python
kuadrat(-1)
kuadrat(-4)
```

```python
print(kuadrat(-1))
print(kuadrat(-4))
print(kuadrat(5))
```

```python
x = [-2, -1, 0, 1, 2, 3]
y = [4, 1, 0, 1, 4, 9]

plt.plot(x, y, '.')
```

```python
plt.plot(x, y, label="Persamaan kuadrat")
plt.title("Gambar grafik x^2", color="red")
plt.xlabel("Nilai x")
plt.ylabel("Nilai y")
plt.legend(loc="lower right")
```

```python
x = np.linspace(-10, 10)
f1 = x**8 + 2*x + 5
f2 = x**2 + 15*x -20

plt.plot(x, f1, label="Persamaan kubik")
plt.plot(x, f2, label="Persamaan kuadrat")
plt.title("Grafik fungsi")
plt.legend(loc="upper left")
```
