2. Jelaskan fungsi dari double asterisk (**) pada sebuah parameter fungsi.
>Double asterisk berfungsi sebagai tanda bahwa suatu *keyword arguments* memiliki jumlah yang tidak diketahui. Keyword arguments merupakan tipe argumen yang nilainya terikat pada sebuah *key*. Secara umum, argumen ini biasa ditulis sebagai `**kwargs`, namun dapat ditulis dalam bentuk lain.


```python
def func1 (**kwargs):
  for key, value in kwargs.items():
    print(f"Keyword dari argumen adalah: {key},")
    print(f"sedangkan value dari argumen adalah: {value}\n")

func1(nama="Khaelano", prodi="TIF", univ="UB")
func1(nama="Budi", prodi="Kedokteran", univ="UB", nim="05329xxxx010352", )
```

3. Jelaskan perbedaan dari keyword `pass` dan `return None`.
>Keyword `pass` memungkinkan sebuah fungsi untuk tidak memiliki kode, biasanya digunakan sebagai placeholder untuk menghindari error. Sedangkan `return None` menunjukkan bahwa suatu fungsi tidak mengembalikan apa-apa.

```python
def func_pass ():
  pass

def func_none():
  return None

func_pass()
func_none()
```
