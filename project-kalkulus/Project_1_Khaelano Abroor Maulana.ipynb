## **KESIMPULAN**

Setelah Anda mencoba semua percobaan di atas, JENIS apasajakah yang dapat dioperasikan pada Sympy? dan jelaskan berbagai macam sintaks yang ada! (jawab dengan menggunakan code)

```
# Sympy dapat mengoperasikan berbagai jenis bilangan seperti:
# Bilangan bulat (Integer), bilangan rasional (Rational), bilangan desimal dengan koma (Float), bilangan imajiner (I), dll.

from sympy import Rational    # from sympy import * dapat digunakan untuk mengimport seluruh class sympy

# Sympy akan secara otomatis mengubah suatu bilangan menjadi bentuk lainnya jika diperlukan (untuk melakukan operasi)
rational = 5 + Rational(3, 5)  # bisa di tulis juga sebagai 5 + S(3)/5
print(type(rational))
# Pada contoh di atas sympy secara otomatis merubah integer 5 menjadi tipe Rational untuk melakukan operasi penjumlahan
```

```
from sympy import sqrt, Integer

# Operasi bilangan bulat (Integer)
integer1 = Integer(4)
integer2 = Integer(9)
integer3 = Integer(6)
integer4 = Integer(27/7)

print(integer4) # Secara otomatis dibulatkan ke bawah

# Bilangan bulat milik sympy (Integer) berbeda dengan bilangan bulat milik python (int)
print(type(integer1))
print(type(int(98)))
print()

# Penjumlahan dan pengurangan
penjumlahan = integer1 + integer2 - 9   # int 9 secara otomatis akan dikonversi menjadi Integer oleh sympy
print(f"Hasil dari operasi tersebut adalah: {penjumlahan} \n")

# Perkalian dan pembagian
perkalian = integer1 * integer3 / integer2
print(f"Hasil dari operasi tersebut adalah: {perkalian} \n")

# Kuadrat dan akar
kuadrat = integer1**2
print(f"Hasil kuadrat: {kuadrat}")
print(f"Tipe kuadrat: {type(kuadrat)}")

akar = sqrt(integer2)
print(f"Hasil akar: {akar}")
print(f"Tipe akar: {type(akar)}")
```

```
from sympy import Rational, pi

# Operasi bilangan rasional
rasional1 = Rational(5,2)  # Membuat suatu bilangan rasional dengan numerator 5 dan denominator 2
rasional2 = Rational(1, 3) # Membuat suatu bilangan rasional dengan numerator 1 dan denominator 3
rasional3 = Rational(9,0)  # Membuat suatu bilangan rasional dengan numerator 9 dan denominator 0

print(f"tipe dari rasional3 adalah: {type(rasional3)} \n")
# Karena 9/0 sama dengan tak hingga, maka sympy secara otomatis mengubah rasional3 menjadi infinity

# Penjumlahan dan pengurangan
penjumlahan = rasional1 + rasional2 + 5 # Angka 5 secara otomatis dikonversi menjadi Rational oleh sympy
print(f"hasil dari operasi tersebut adalah: {penjumlahan} \n")

# Perkalian dan pembagian
perkalian = rasional1 / 8 * rasional3
print(f"hasil dari operasi tersebut adalah: {perkalian}")
print(f"tipe data: {type(perkalian)} \n")

# Konversi menjadi Rational
pi_dalam_pecahan = Rational(Float(pi)) # Disini terdapat dua kali konversi yaitu pi -> Float -> Rational
print(f"Pi dalam pecahan: {pi_dalam_pecahan}")

# Method limit_denominator berfungsi untuk menetapkan batas maksimal denominator
print(f"Pi dalam pecahan dalam bentuk sederhana: {pi_dalam_pecahan.limit_denominator(10)} \n")
```

```
from sympy import Float

# Operasi bilangan desimal berkoma
float1 = Float(10.6)
float2 = Float(7.6)
float3 = Float(3.5555555555555, 3)  # Argumen kedua akan memulatkan dan membatasi lebar float menjadi 3 digit

# Penjumlahan dan pengurangan
penjumlahan = float1 + float3 - 10.0  # 10 dikonversi menjadi Float
print(f"Hasil dari operasi tersebut adalah: {penjumlahan} \n")

# Perkalian dan pembagian
perkalian = float1 * 2.5 * float2
print(f"Hasil dari operasi tersebut adalah: {perkalian}")
```

```
from sympy import S, I, Q, ask

# Class S dapat digunkan untuk memepermudah kita dalam mendefinisikan sesuatu
print(S.Infinity) # S.Infinity sama dengan Infinity
print()

print(S.One)   # S.One sama dengan 1
print(S.Zero)  # S.Zero sama dengan 0
print()

print(S.ImaginaryUnit) # S.ImaginaryUnit sama dengan I (imajiner)
print()

print(S.Pi) # S.Pi sama dengan pi
print()

# Bilangan kompleks
kompleks = 3 + 4*I   # Otomatis terkonversi menjadi compleks
ask(Q.complex(kompleks))

# Penggunaan ask dan Q untuk mengetahui jenis bilangan
integer = 98
print(ask(Q.odd(integer)))
print(ask(Q.imaginary(integer)))
print(ask(Q.prime(integer)))
print(ask(Q.irrational(integer)))
print(ask(Q.composite(integer)))
```
